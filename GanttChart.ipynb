```python
import matplotlib.pyplot as plt
import pandas as pd

# Project timeline data
tasks = [
    'Requirement Gathering & Planning',
    'Wireframes & Prototypes Design',
    'Backend Development & Setup',
    'Frontend Development & UI Enhancements',
    'Comprehensive Testing & Debugging',
    'Deployment & Final Review',
    'Maintenance & Updates'
]

# Define start weeks and durations in weeks
start_weeks = [0, 1, 2, 5, 8, 10, 11]
durations_weeks = [1, 1, 3, 3, 2, 1, 1]  # Duration of each task in weeks

# Creating a dataframe for easy plotting
df = pd.DataFrame({
    'Task': tasks,
    'Start': start_weeks,
    'Duration': durations_weeks
})

# Calculate end weeks
df['End'] = df['Start'] + df['Duration']

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Define bar width and colors
bar_width = 0.5  # Keep the bar width consistent
colors = ['#6A5ACD', '#00BFFF', '#3CB371', '#FF8C00', '#FF69B4', '#FFD700', '#7B68EE']

# Plot each task
for i, task in enumerate(df['Task']):
    ax.barh(task, df['Duration'][i], left=df['Start'][i], height=bar_width, color=colors[i % len(colors)])

# Adjust x-ticks for weeks 0-11
ax.set_xticks(range(0, 12))  # Set x-ticks for weeks 0-11
ax.set_xticklabels([f'Week {i}' for i in range(12)])  # Label x-ticks with weeks

# Labels and title
ax.set_xlabel('Timeline (Weeks)', fontsize=12)
ax.set_ylabel('Tasks', fontsize=12)
ax.set_title('Project Timeline Gantt Chart', fontsize=14)

# Enhancing the grid and layout
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, color='grey')
plt.tight_layout()

# Display the chart
plt.show()
```
